```python
#drop columns we dont need for client requisitions
amz.drop(columns=['director','description', 'rating', 'director'], inplace=True)
dis.drop(columns=['director', 'description', 'rating', 'director'], inplace=True)
hu.drop(columns=['director', 'description', 'rating', 'director'], inplace=True)
net.drop(columns=['director', 'description', 'rating', 'director'], inplace=True)
```

```python
#fill null values
net.fillna('no data', inplace=True)
amz.fillna('no data', inplace=True)
dis.fillna('no data', inplace=True)
hu.fillna('no data', inplace=True)
```

```python
#add column that id's what platform each movie/tvshow is from
amz['platform'] = 'amazon prime'
net['platform'] = 'netflix'
hu['platform'] = 'hulu'
dis['platform'] = 'disney plus'
```

```python
#split duration into two columns, discriminating by string contents: either 'min' or 'seaons'
amz['movie_duration_mins']=amz['duration'].apply( lambda i: i.split()[0] if i.split()[1]=='min' else '0')
amz['seasons']=amz['duration'].apply( lambda i: i.split()[0] if i.split(" ",1)[1]=='Seasons' or i.split(" ",1)[1]=='Season' else '0')
```

```python
net['movie_duration_mins']=net['duration'].apply( lambda i: i.split()[0] if i.split()[1]=='min' else '0')
net['seasons']=net['duration'].apply( lambda i: i.split()[0] if i.split(" ",1)[1]=='Seasons' or i.split(" ",1)[1]=='Season' else '0')
```

```python
hu['movie_duration_mins']=hu['duration'].apply( lambda i: i.split()[0] if i.split()[1]=='min' else '0')
hu['seasons']=hu['duration'].apply( lambda i: i.split()[0] if i.split(" ",1)[1]=='Seasons' or i.split(" ",1)[1]=='Season' else '0')
```

```python
dis['movie_duration_mins']=dis['duration'].apply( lambda i: i.split()[0] if i.split()[1]=='min' else '0')
dis['seasons']=dis['duration'].apply( lambda i: i.split()[0] if i.split(" ",1)[1]=='Seasons' or i.split(" ",1)[1]=='Season' else '0')
```

```python
#concatenate all dfs into one in order to run querys
content = pd.concat([amz, net, hu, dis], axis=0)
```

```python
content.drop(columns=['duration'], inplace=True)
```

```python
content.sample(10)
```

```python
content.dtypes
```

```python
#change dtype to int for seasons and duration (mins)
content = content.astype({'seasons' : 'int'})
content = content.astype({'movie_duration_mins' : 'int'})
```
